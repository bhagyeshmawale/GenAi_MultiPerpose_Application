```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# Step 1: Fetch the HTML content (optional but recommended)
url = 'https://deepmind.google/technologies/gemini/pro/'
response = requests.get(url)
html_content = response.text
```

```python
# Step 2: Parse the HTML content using BeautifulSoup (optional)
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
# Step 3: Extract tables using pandas
tables = pd.read_html(str(soup)) 
```

```python
# Step 4: Select the table you need
# Assuming you want the first table (index 0)
df = tables[0]
```

```python
# Step 5: Inspect the DataFrame
df.fillna('')
```
